```python
##########################################################################
######################### BACK TO HOGWARTS ###############################
##########################################################################

"""
    DocString :
    
    A) Introduction :
    This game took the motif from Harry Potter(2001 - 2011),but is set in Hogwarts
    a few years after Harry Potter has won his final battle against Voldemort.
    This game consists of three stages.Players must possess a bit of
    Harry Potter knowledge from the movie series.
    
    Stage 1 : Decide opeing Dean's letter with intro scenario
    Stage 2 : Solve quizes to join the club
    Stage 3 : Duel battle with random students
    
    B) Known bugs and/or Errors :
    None.


"""




### color, time, random ###
from colorama import Fore
import time
import sys
from sys import exit
import random


###########################################################################
#####################           Game Start         ########################
##################### Intro_background_description ########################
###########################################################################

print(f"""
        {'='*20} Welcome to BACK TO HOGWARTS {'='*20}
        You are about to start "Back To Hogwarts".
        Please answer below quetions before you start the game.
        {'='*70}
    """)
time.sleep(2)

def intro_game(count):
    global user_name
    global owl_name
    global dean_name
    
    count += 1
    
    ###user_input_intro_game###
    user_name = input(prompt = """What nickname do you want to use in the game? \n""")
    owl_name = input(prompt = "What's your family owl's name? \n")
    dean_name = input(prompt = "What's your school dean's name? \n")
    
    ### Make small letter to compare user's answer ###
    user_name = user_name.casefold()
    owl_name = owl_name.casefold()
    dean_name = dean_name.casefold()

    ###Code for input difference// No same name ###
    if user_name == owl_name or user_name == dean_name or owl_name == dean_name:
        
        ## Chance to retype once => Retrun to input
        if count != 1 :
            print(
                "\033[1m"
                f"""{Fore.RED}Name's can't be the same.\t
please retype the names to start "BACK TO HOGWARTS"{Fore.RESET}\n"""
                "\033[0m")
            intro_game(0)
            
        # Fail the game
        else :
            print(
                "\033[1m"
                f"""{Fore.RED} OOPS! Same agin! \t
                please restart "BACK TO HOGWARTS" game.{Fore.RESET}\n"""
                "\033[0m")
            fail_1()
    
    ### Game start!
    else:
        print("\033[1m"
              f"""{Fore.MAGENTA}Awesome! Let's get started "BACK TO HOGWARTS"!!{Fore.RESET}\n"""
              "\033[0m")
        print("""
        
        
        
        
        
        
        
        """)
        get_ready()


###################### Stage 1 : Intro_scenario_back to Hogwarts ##########################
def get_ready():
    
    ##Stage 1 scenario
    print(f"""
    Mom : WAKE UP {user_name.capitalize()}!!! YOU WILL MISS YOUR TRAIN.
       

    " Hi, I'm {user_name.capitalize()}. I’m a 4th year student at Hogwarts.
    The legend of Harry Potter winning the battle against “You-Know-Who” is now widespread. 
    Since Harry Potter has won the battle, our world has been peaceful until… "
    

    "Hoot Hoo Hoot Hoo"

    Mom : Oh, {owl_name.capitalize()} is comming. {user_name.capitalize()}, can you get mails?
    {user_name.capitalize()} : Good job {owl_name.capitalize()}. Hummm, I got a mail from a professor!
    """)
    
    ##First_game_selection : wrong choice go back to the selection##
    ## chance only 3 ###
    chance = 2

    while chance <= 2:
        first_selec = input(prompt =f"""
                {'-'*60}
                You received a mail which looks important from Dean.
                Do you want to open it?
                    A. Yes
                    B. No
                {'-'*60}\n
        """)
        first_selec = first_selec.casefold()
        

        # Chose 1 => Open the dean's letter
        if 'a' in first_selec or 'y' in first_selec :
            print(f"""
        Dear.{user_name.capitalize()}

         We are so excited for a new school year  with you.
         I hope that you have prepared all of your equipment for school.
         Also, we'd recommend that you join the "PHOENIX" club this year.
         Please notify us upon your arrival. 
         Looking forward to seeing you soon.
                                                        Dean, Prof. {dean_name.capitalize()}

        """)
            
            break

        # choose 2
        elif 'b' in first_selec or 'n' in first_selec :
            
            ## Fail the game
            if chance == 1:
                print("\033[1m" f""" {Fore.BLUE}
 {user_name.capitalize()} : Oh My God. I missed my train!!!{Fore.RESET}""" "\033[0m")
                fail_2()
                
            ## Return to input_first_selec
            else:
                chance -= 1
                print("\033[1m" f"""{Fore.RED}
            {'*'*70}
            OOPS! You might miss important information. You have {chance} chance left.
            {'*'*70}{Fore.RESET}\n"""
            "\033[0m")


        ## Mistype option
        else:
            
            ## Fail the game due to run out of chances
            if chance == 1:
                fail_1()
            
            ## Return to input_first_selec
            else:
                chance -= 1
                print("\033[1m" f"""{Fore.RED}
                {'*'*60}
                You are not able to go back to Hogwart if you mistype.\t
                You have {chance} chance left.
                {'*'*60}{Fore.RESET}\n"""
                "\033[0m")
               

    ## End intro_scenario ###
    print(f"""
    {user_name.capitalize()} : This is weird.
        The prof.{dean_name.capitalize()} has never asked me to join a club before. 
        I should check in with my friends once I get to school.\n
    """)
    
    p_enter = input(prompt = "Press Enter to go next stage. \n")
    print("""
    
    
    
    
    
    
    
    """)
    sch_join()


############################## Stage 2 : Join the club "PHOENIX"  ##############################


def sch_join():

    ### input Best Friend name ###
    print( f" ??? : Hey, {user_name.capitalize()}! Have you received the mail from the Prof?")
    bf_name = input(prompt="What is your Best Friend's name? ")
    print(f"{user_name.capitalize()} : Yes {bf_name.capitalize()}, I wonder what this is all about.\
Do you have any idea about the club?")

    
                
    ### Stage 2 scenario ###
    print(f"""
    Prof.{dean_name.capitalize()} : HI all, I’m sure you are all wondering\
why I sent a letter inviting all of you to join this club.
        Unfortunately, unknown objects that carry “You-know-who”’s power have been sneaking into the school.
        your mission is to join "PHOENIX" to prepare yourselves against the uncertain future. 

                  
    {user_name.capitalize()} : Professor {dean_name.capitalize()}, then what is our first task?
    
    Prof. {dean_name.capitalize()} : Great question. First, you must take a test to see whether you are qualified.
        Answer these following questions. Fingers crossed for you :)
        """)
 
    
    ### 1st quiz ###
    chance = 2
    
    while True : 
        second_selec = input(prompt = f"""\n
                {'-'*50}
                Who is the auther of Harry Potter?
                    A. J.R.Tolkien
                    B. J.K.Rowling
                    C. George.R.Martin
                    D. I don't know (Hint hide)
                {'-'*50}\n
        """)
        second_selec = second_selec.casefold()

        
        ### Chose 2 ###
        if 'b' in second_selec or 'k' in second_selec :
            print (f"""
    Prof.{dean_name.capitalize()} : That is correct! Answer the next question!!""")
            break
        
        ### Chose 4 ###
        elif 'd' in second_selec or 'don\'t' in second_selec :
            
            ### Fail the game ###
            if chance == 1 :
                print("\033[1m" f""" {Fore.BLUE}
    Prof.{dean_name.capitalize()} : {user_name.capitalize()}, you need improvement.{Fore.RESET}""" "\033[0m")
                fail_2 ()
                
            ### Return to second_selec input ###    
            else :
                chance -= 1
                
                ###Hint
                print (f"""
                Look at the hint below :)
                J.R.Tolkien : A renowned writer of the “The Road of Ring” series.
                J.K.Rowling : A renowned writer of the “Harry Potter” series.
                George.R.Martin : A renowned writer of the “The Game of Throne” series.
                """)
                print("\033[1m" f"""{Fore.RED}
            {'*'*40}
            Does it help? You have {chance} chance left.
            {'*'*40}{Fore.RESET}\n"""
            "\033[0m")
                
                
        ### chose 1,3 or mistype ###
        else :
            ### Chose 1 or 3 ###
            if 'a' in second_selec or 'c' in second_selec or 'r' in second_selec or 'g' in second_selec :
                print("\033[1m" f""" {Fore.BLUE}
    Prof.{dean_name.capitalize()} : {user_name.capitalize()}, you need improvement.{Fore.RESET}""" "\033[0m")
                fail_2 ()
            
            ### Mistype option ###
            else :
                
                ## Fail the game
                if chance == 1 :
                    fail_1()
                
                ## Return to second_selec input ##
                else :
                    chance -= 1
                    
                    print("\033[1m" f"""{Fore.RED}
                {'*'*50}
                Time rus out while you were mistyping!\t
                You have {chance} chance left.
                {'*'*50}{Fore.RESET}\n"""
                "\033[0m")
                    
        
    ###### 2nd Quiz ###
    while chance <= 2:
        
        third_selec = input(prompt = f"""
                {'-'*40}
                {bf_name.capitalize()} : This one is tough.
                
                Who is Harry Potter's friend? 
                    A. Draco Malfoy
                    B. Sirus Black
                    C. Ron Weasley
                    D. I don't know
                {'-'*40}\n
        """)
        third_selec = third_selec.casefold()

         ### Chose 1 => Able to join the club###
        if 'c' in third_selec or 'ron' in third_selec :
            print (f"""
    Prof.{dean_name.capitalize()} : Good job {user_name.capitalize()}!! Welcome to the club""")
            break

        ### Chose 2 ###
        elif 'd' in third_selec or 'don\'t' in third_selec :
            
            ### Fail the game ###
            if chance == 1 :
                fail_1 ()

            ### getting another chance ###    
            else :
                
                ## Hint
                chance -= 1
                print (f"""
                Look at the hint below :)
                Harry Potter went to Hogwarts with Ron Weasley and Hermione Granger.\n
                """)
                print("\033[1m" f"""{Fore.RED}Does it help? You have {chance} chance left.{Fore.RESET}""" "\033[0m")


        ### chose 1,3 or mistype ###
        else :
            
            ### Chose 1 or 3 ###
            if 'a' in third_selec or 'b' in third_selec or 'draco' in third_selec or 'ron' in third_selec :
                print("\033[1m" f""" {Fore.BLUE}
    Prof.{dean_name.capitalize()} : Oh... {user_name.capitalize()}, You just missed golden opportunity.\n{Fore.RESET}""" 
                      "\033[0m")
                fail_2 ()

            ### Mistype ###
            else :
                
                # Fail the game
                if chance == 1 :
                    fail_1()

                # Return to third_select input
                else :
                    chance -= 1
                    print("\033[1m" f"{Fore.RED}You won't able to join the club if you type wrong answer.\
                    Please re-type from the options. you have {chance} chance left.{Fore.RESET} \n""\033[0m")


    p_enter = input(prompt = "Press Enter to go next stage. \n")
    print("""
    
    
    
    
    
    
    
    
    
    """)
    club_acti()


################################# Stage 3 : Club activity - Duel battle #####################################


def club_acti () :
    print(f"Prof.{dean_name} : Welcom to the club")
    
    ##input club leader name & scenario ##
    club_leader = input(prompt = "What is club leader's name?\n ")
    
    #Stage 3 Scenario
    print(f"""
    {club_leader} : I'm the leader of the club. As you know, the dark magic is getting powerful.
            As members of “aaa,” we must find why the dark power has been growing. 
            Also, through duels, we are going to learn how self-defense against dark magic.
            Your mission is get high score by winning duels against four different students. 
            If you lose, you will lose your club membership. 

    """)
    
    ### variable for next def ##
    count = 4
    duel_game(count)

user_lst = []
random_lst = []
random_lst_count = []
user_lst_count =[]

### self def = function loop ###
def duel_game (count) :
    
    ### Stop the deul_game
    if count == 0 :
        
        print(f"""
        DUEL RESULT
        """)
        print(f"{user_name} : ", user_lst)
        print("Competitor :", random_lst)
        
        
        ## Count 'Attact' for each to compare    
        count_user_result = user_lst_count.count('attack')
        count_random_result = random_lst_count.count('attack')
            
       
        ## user win the game ##
        if count_user_result > count_random_result :
            
            return complete_game ()
        
        ## user lost the game##
        else :
            print("\033[1m" f"""{Fore.BLUE}
    {user_name.capitalize()} : I can't believe I lost the duel...{Fore.RESET}
            """ "\033[0m")
            
            return fail_2()

    count -=1

    ### Pick random duel competitor ###
    duel_stuents = ['Dave', 'Jane', 'Tom', 'Racheal', 'Andy']
    random_student = random.choice(duel_stuents)

    print(f"""
    {random_student} : Saying sorry in advance.
    {user_name.capitalize()} : HA! Will see who is going to be sorry.
    """)

    #### Pick competitor's random spell ###
    duel_lst = [['stupefy','Imperio','Reducto'],['episkey','expelliarmus','protego']]
    randnum = random.choice(duel_lst)
    randspell = random.choice(randnum)

    ##Distinguish spell 'Attack' or 'Defense'##
    if randspell in duel_lst[0] :
        competitor = 'attack'

    elif randspell in duel_lst[1]:
        competitor = 'defense'


    ### input user's magic spell ###
    chance = 2
    chance_1 = 2
    while chance <=2 :

        user_choice = input(prompt = f"""
                {'-'*40}
                Choose attack or defense.
                    A. attack            
                    B. defense           
                {'-'*40}
            \n""")
        user_choice = user_choice.casefold()

    ### Chose A    
        if 'a' in user_choice:
            chance = 4
            
            user_choice = 'attack'
            
            ## input attack_magic
            while chance_1 <=3 :
                
                ##chose among 3 attack magic
                attack_magic = input(prompt = f"""
                {'-'*40}
                 What attack magic do you want to use for duel? 
                     A. Stupefy
                     B. Imperio
                     C. Reducto
                {'-'*40}
                \n """)
                attack_magic = attack_magic.casefold() ####String option###
            
                ## define attack_magic variable with user selection
                if 'a' in attack_magic or 's' in attack_magic :
                    attack_magic = "Stupefy"
                    break

                elif 'b' in attack_magic or 'i' in attack_magic :
                    attack_magic = "Imperio"
                    break


                elif 'c' in attack_magic or 'r' in attack_magic :
                    attack_magic = "Reducto"
                    break

                ## Mistype option ###
                else :

                    ## Fail the game
                    if chance_1 == 1 :

                        fail_1 ()

                    ## Return to Attack magic input 
                    else :
                        chance_1 -= 1
                        chance -= 1
                        print("\033[1m" f"{Fore.RED}You won't able to use Attack magic if you type wrong answer.\
Please re-type from the options. you have {chance} chance left.{Fore.RESET} \n""\033[0m")
                        
        ### Chose B
        elif 'b' in user_choice or 'd' in user_choice :
            user_choice = 'defense'
            chance = 4

            while chance_1 <=3 :

                ### Choose among 3 defense magic
                defense_magic = input(prompt = f"""
                {'-'*40}
                 What defense magic do you want to use for duel?
                     A. Episkey
                     B. Expelliarmus
                     C. Protego
                {'-'*40}
                """)
                defense_magic = defense_magic.casefold()
                
               ## define Defense_magic variable with user selection 
                if 'a' in defense_magic or 'sk' in defense_magic :
                    defense_magic = "Episkey"
                    break

                elif 'b' in defense_magic or 'mus' in defense_magic :
                    defense_magic = "Expelliarmus"
                    break

                elif 'c' in defense_magic or 't' in defense_magic :
                    defense_magic = "Protego"
                    break

                ## Mistype option ##
                else :

                    ## Fail the game
                    if chance ==_1 :
                        fail_1()

                    ## return to Defense magic input
                    else :
                        chance_1 -= 1
                        chance -= 1
                        print("\033[1m" f"{Fore.RED}You won't use Defense magic if you type wrong answer.\
Please re-type from the options. you have {chance} chance left.{Fore.RESET} \n""\033[0m")

        ## Mistype user_choice_Q
        else :
            if chance == 1 :
                fail_1()


            else :
                chance -= 1
                print("\033[1m" f"{Fore.RED}You won't able to win the duel if you type wrong answer.\
Please re-type from the options. you have {chance} chance left.{Fore.RESET} \n""\033[0m")



    ## Competitor using magic list
    random_lst.append([competitor,randspell])
    random_lst_count.append(competitor)
    user_lst_count.append(user_choice)


    ## Tie the duel
    if user_choice == competitor:
        
        ## User chose Attack
        if 'a' in user_choice :
            
            ## list to compare result
            user_lst.append([user_choice,attack_magic])
            
            
            print("\033[1m" f"""
    {user_name.capitalize()} : {attack_magic.capitalize()}!
    {random_student} : {randspell}!
            """ "\033[0m")
    
        # User chose Defense
        else :
            user_lst.append([user_choice,defense_magic])
            
            print("\033[1m" f"""
    {user_name.capitalize()} : {defense_magic}!
    {random_student} : {randspell}!
            """ "\033[0m")
        
        
        print (f"""
    Prof.{dean_name.capitalize()} : Hummmmmm.... It's hard to tell who won the game. \t
    Let's rematch the game with another student!
    {random_student} : I'm goint to win next time.
    {user_name.capitalize()} : It's a pie in the sky!!
    """)
        duel_game(count)



    ## User win or loose the duel 
    else :
        
        ##User win the duel
        if 'a' in user_choice :
            user_lst.append([user_choice,attack_magic])
            

            print("\033[1m" f"""
    {user_name.capitalize()} : {attack_magic}!
    {random_student} : {randspell}!""" "\033[0m")
            print(f"""
    {user_name.capitalize()} : HAHA! I guess someone needs more practice to join the club
            {Fore.RED}{'-'*30}
                !You won the duel! 
            {'-'*30}{Fore.RESET}
            """)
            print("""
            
            
            
            
            """)

            duel_game(count)

        #User loose the duel
        else :
            user_lst.append([user_choice,defense_magic])
            
            
            print("\033[1m" f"""
    {user_name.capitalize()} : {defense_magic}!
    {random_student} : {randspell}!""" "\033[0m")
            print(f"""
    {random_student} : HAHAHA! What I told you! You are no match for me!!
            {Fore.RED}{'-'*30}
                !You lost the duel!
            {'-'*30}{Fore.RESET}
            """)
            print("""
            
            
            
            
            """)

            duel_game(count)


################################ Complete game ###############################
def complete_game ( ) :
    print("""
    
    
    
    
    
    
    
    
    """)
    
    
    print("\033[1m" f"""{Fore.GREEN}
    {user_name.capitalize()} : Yes!!! I won the duel!
    Now, I feel confidece about myself!!
    I think I'm ready to defense myself even though dark power comes!!
    {'='*70}
    
            ____      _      __  __  U _____ u    U _____ u _   _    ____    
        U /"___|uU  /"\  uU|' \/ '|u\| ___"|/    \| ___"|/| \ |"|  |  _"\   
        \| |  _ / \/ _ \/ \| |\/| |/ |  _|"       |  _|" <|  \| |>/| | | |  
         | |_| |  / ___ \  | |  | |  | |___       | |___ U| |\  |u | |_| |\ 
          \____| /_/   \_\ |_|  |_|  |_____|      |_____| |_| \_|  |____/ u 
          _)(|_   \\    >><<,-,,-.   <<   >>      <<   >> ||   \\,-.|||_    
         (__)__) (__)  (__)(./  \.) (__) (__)    (__) (__)(_")  (_/(__)_)   

    {'='*70}{Fore.RESET}
    """ "\033[0m")
    
    
    start_over_fianl ()


################################ Fail : Use 3 chances #############################
def fail_1():
    print("""
    
    
    
    
    
    
    
    
    """)
    
    print(f"""{Fore.BLUE}{'='*70}
    OOPS! I'm sorry but you used all your chances.
                    
           ____      _      __  __  U _____ u       U  ___ u__     __ U _____ u   ____     
        U /"___|uU  /"\  uU|' \/ '|u\| ___"|/        \/"_ \/\ \   /"/u\| ___"|/U |  _"\ u  
        \| |  _ / \/ _ \/ \| |\/| |/ |  _|"          | | | | \ \ / //  |  _|"   \| |_) |/  
         | |_| |  / ___ \  | |  | |  | |___      .-,_| |_| | /\ V /_,-.| |___    |  _ <    
          \____| /_/   \_\ |_|  |_|  |_____|      \_)-\___/ U  \_/-(_/ |_____|   |_| \_\   
          _)(|_   \\    >><<,-,,-.   <<   >>           \\     //       <<   >>   //   \\_  
         (__)__) (__)  (__)(./  \.) (__) (__)         (__)   (__)     (__) (__) (__)  (__) 

    {'='*70}{Fore.RESET}
    """)

    
    start_over_fianl()

################################# Fail : Lost the game ######################    
def fail_2() :
    print("""
    
    
    
    
    
    
    
    
    """)
    
    print(f"""
    {Fore.BLUE}{'='*70}
                    
           ____      _      __  __  U _____ u       U  ___ u__     __ U _____ u   ____     
        U /"___|uU  /"\  uU|' \/ '|u\| ___"|/        \/"_ \/\ \   /"/u\| ___"|/U |  _"\ u  
        \| |  _ / \/ _ \/ \| |\/| |/ |  _|"          | | | | \ \ / //  |  _|"   \| |_) |/  
         | |_| |  / ___ \  | |  | |  | |___      .-,_| |_| | /\ V /_,-.| |___    |  _ <    
          \____| /_/   \_\ |_|  |_|  |_____|      \_)-\___/ U  \_/-(_/ |_____|   |_| \_\   
          _)(|_   \\    >><<,-,,-.   <<   >>           \\     //       <<   >>   //   \\_  
         (__)__) (__)  (__)(./  \.) (__) (__)         (__)   (__)     (__) (__) (__)  (__) 

    {'='*70}{Fore.RESET}
    """)

    
    start_over_fianl()

############################### Game restart ###################################
def start_over_fianl(): 
    start_over_q = input(prompt="""Do you want to restart the game? 
    A. Yes, I want to start over
    B. No, I'm done
    """)
    start_over_q = start_over_q.casefold()
    
    ## Restart the game
    if 'a' in start_over_q or 'y' in start_over_q :
        p_enter = input(prompt = "Press enter to start over")
        print(f"""
        
        
        
        
        
        """)
        intro_game(1)
    
    ## Bye Bye game
    else:
        print("\033[1m" f"""
        {Fore.YELLOW}{'-'*60}
        Thans you so much for playing the game. Hope to see you soon agian!{Fore.RESET}
        {'-'*60}{Fore.RESET}
        """ "\033[0m")
        sys.exit()


intro_game(1)
```
